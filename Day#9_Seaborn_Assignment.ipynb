## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import seaborn as sns
sns.set() # to set the style
```

# Q9: Draw barplot represents survived people based on sex

```python
sns.barplot(x='survived', y='sex',data=df); # it shows the average 
```

# Q10: Draw catplot represents survived people based on embarked

```python
sns.catplot(
    data=df, # Data
    kind="count", # Chart to use
    x="embarked", # X axis
    hue="survived", # create bar for each hue
    height=5
)
plt.title("Survived people based on embarked");
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
plt.figure(figsize=(15,8))

sns.boxplot(
    x="age", 
    y="sex",
    hue="pclass",
    data=df)

plt.title("Distribution of male and female based on age and pclass");
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
df_hm = df[['sibsp','parch','age','fare','survived']]
df_hm
```

```python
df_hm.corr()
```

```python
sns.heatmap(df_hm.corr(), annot = True, fmt='.01g');
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
sns.catplot(
    data=df, # Data
    kind="point", # Chart to use
    x="sibsp", 
    y ='survived');
```

# Q14: Draw extra insights [Optional]

```python
sns.relplot(
    x="age", 
    y="fare", 
    hue="survived",
    size="pclass",
    palette="muted",
    height=6, 
    data=df
)

plt.title("People survival based on age, fare, and class");
```

> We can see young people in higher class tend to have a higher survival rate than older people.
